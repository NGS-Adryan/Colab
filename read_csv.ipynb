```python
import pandas as pd

URL_CSV = 'https://raw.githubusercontent.com/alura-cursos/formacao-data-science/master/movies.csv'
csv = pd.read_csv(URL_CSV).set_index('movieId')
csv.head(30)
```

```python
csv['genres'] = [', '.join(genre.split('|')) for genre in csv['genres']]
csv.head(30)
```

```python
csv = csv[csv['genres'].str.contains('Comedy', case=False, na=False)]
csv['year'] = csv['title'].str.split('(', n=1).str[1].str.rstrip(')')
csv['title'] = csv['title'].str.replace(r'\(\d{4}\)', '', regex=True).str.strip()

csv.reset_index(drop=True)
```

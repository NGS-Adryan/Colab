```python
# -*- coding: utf-8 -*-

import requests
from bs4 import BeautifulSoup
from time import sleep

def get_price(tiker: str):
  tiker = tiker.upper()
  URL = f'https://www.google.com/finance/quote/{tiker}:BVMF?hl=pt'
  SELECTORS = {
    'CSS_SELECTOR_NAME': 'div.zzDege',
    'CSS_SELECTOR_PRICE': 'div.YMlKec.fxKbKc'
  }
  response = requests.get(URL)
  soup = BeautifulSoup(response.text, 'html.parser')
  name, price = (soup.select_one(SELECTORS['CSS_SELECTOR_NAME']).text,
                 soup.select_one(SELECTORS['CSS_SELECTOR_PRICE']).text)
  return [name, str(price).replace('\xa0', ' ')]


get_price('taee11')
```

```python
tikers = ['bbas3', 'taee11']
price_quotations = {'Tiker': [], 'Nome': [], 'Preço': []}
while True:
  for tiker in tikers:
    name, price = get_price(tiker)
    price_quotations['Tiker'] += [tiker.upper()]
    price_quotations['Nome'] += [name]
    price_quotations['Preço'] += [price]

  print(price_quotations)
  # sleep(5)
  break
  if input('Pegar cotações novamente? ') != 's':
    break

```

```python
import pandas as pd

df = pd.DataFrame(price_quotations)
df
```

```python
rows = ws.get_all_values()
# for row in rows:
#  print(row)
```

```python
import pandas as pd

df = pd.DataFrame.from_records(rows)
df.columns = df.iloc[0]
df = df[1:]
df.reset_index(drop=True)
```

```python
from geopy.geocoders import Nominatim

latitude = -30.00099
longitude = -51.18977

geolocator = Nominatim(user_agent="Adryan")
location = geolocator.reverse((latitude, longitude), language='pt-BR')
location.address
```

```python
import requests

endereço = []

for lat, lon in zip(df['Latitude'], df['Longitude']):
  # print(lat, '', lon)
  lat, lon  = float(lat.replace(',', '.')), float(lon.replace(',', '.'))
  url = f'https://nominatim.openstreetmap.org/reverse?format=json&lat={lat}&lon={lon}&language=pt-BR'
  response = requests.get(url)
  data = response.json()
  try:
    endereço.append(data['address']['road'])
  except:
    endereço.append(data['address'])
  #location = geolocator.reverse((lat, lon), language='pt-BR')
  #endereço.append(location.address)
  #endereço

df['Endereço'] = endereço

df.head(50)
```

```python
enderecos = df['Endereço'].tolist()

for i in range(len(enderecos)):
  ws.update(f'C{i + 2}', str(enderecos[i]))
```
